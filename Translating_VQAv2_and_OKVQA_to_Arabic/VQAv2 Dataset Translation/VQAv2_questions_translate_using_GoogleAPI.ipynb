# Google Translate API

```python
pip install googletrans==3.1.0a0
```

```python
from googletrans import Translator
translator = Translator()
translation = translator.translate("world", dest='ar')
print(translation.text)
```

# Translating Training  Questions

```python
import json

with open('original-dataset/v2_OpenEnded_mscoco_train2014_questions.json', 'r') as datafile:
  data = json.load(datafile)
```

```python
type(data),
```

```python
data.keys(), len(data['questions']),data['questions'][0]
```

```python
data_1 = data['questions']
```

```python
len(data_1)
```

```python
data_1[0:5]
```

```python
for item in range(0,len(data_1)):
  print(item)
  translation_1 = translator.translate(data_1[item]['question'], dest='ar')
  data_1[item]['question_ar'] = translation_1.text
```

```python
data_1[443750:443757]
```

```python
import json
with open('translated-dataset/v2_OpenEnded_mscoco_train2014_questions_translated.json', 'w') as f:
    json.dump(data_1, f, ensure_ascii = False)
```

# Translating Training Annotations

```python
import json

with open('original-dataset/v2_mscoco_train2014_annotations.json', 'r') as datafile:
  data = json.load(datafile)
```

```python
data.keys()
```

```python
len(data['annotations']),data['annotations'][0]
```

```python
data_1 = data['annotations']
```

```python
len(data_1)
```

```python
data_1[0:5]
```

```python
for item in range(0,len(data_1)):
  print(item)
  for an in range(0,len(data_1[item]['answers'])):
    print(an)
    translation_1 = translator.translate(data_1[item]['answers'][an]['answer'], dest='ar')
    data_1[item]['answers'][an]['answer_ar'] = translation_1.text
```

```python
data_1[69353:69363]
```

```python
import json
with open('translated-dataset/v2_mscoco_train2014_annotations_translated.json', 'w') as f:
    json.dump(data_1, f, ensure_ascii = False)
```
