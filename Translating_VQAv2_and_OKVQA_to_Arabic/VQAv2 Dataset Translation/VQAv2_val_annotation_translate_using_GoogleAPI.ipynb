**Automatic batch translating**

```python
# The translation code working in batches and save the resulted translation into JSON file 'translated-dataset/v2_mscoco_val2014_annotations_translated2.json incrementally

from google.colab import drive
import sys
import torch
# from transquest.algo.sentence_level.siamesetransquest.run_model import SiameseTransQuestModel
import json
import numpy as np
import os
from tqdm import tqdm
import nltk
from nltk.tokenize import word_tokenize
import csv
from googletrans import Translator

# ... (rest of your existing code, including functions: load_dataset, extract_source_target, etc.)

def translate_in_batches(data, batch_size=100, output_file='translated-dataset/v2_mscoco_val2014_annotations_translated2.json'):
    """Translates the dataset in batches and saves the results incrementally."""

    translator = Translator()
    total_items = len(data)
    translated_data = []
    start_index = 0

    # Check for existing translated file
    if os.path.exists(output_file):
      with open(output_file, 'r') as f:
          dataset = json.load(f)
          translated_data = dataset['annotations']
      start_index = len(translated_data)
      print(f"Resuming translation from index {start_index}")


    for i in tqdm(range(start_index, total_items, batch_size), desc="Translating in batches"):
        end_index = min(i + batch_size, total_items)
        batch = data[i:end_index]

        for item in batch:
                translation1 = translator.translate(item['multiple_choice_answer'], dest='ar')
                item['multiple_choice_answer_ar'] = translation1.text
                for an in range(len(item['answers'])):
                    translation2 = translator.translate(item['answers'][an]['answer'], dest='ar')
                    item['answers'][an]['answer_ar'] = translation2.text
        translated_data.extend(batch)

        # Save the translated data incrementally
        with open(output_file, 'w') as f:
            json.dump({'annotations': translated_data}, f, ensure_ascii=False, indent=4)

    print(f"Translation complete. Data saved to {output_file}")
    return translated_data


if __name__ == "__main__":
    drive.mount('/content/drive')
    %cd /content/drive/MyDrive/ColabData/VQAv2

    with open('original-dataset/v2_mscoco_val2014_annotations.json', 'r') as datafile:
        data = json.load(datafile)

    data_1 = data['annotations']

    translated_data = translate_in_batches(data_1)
```
