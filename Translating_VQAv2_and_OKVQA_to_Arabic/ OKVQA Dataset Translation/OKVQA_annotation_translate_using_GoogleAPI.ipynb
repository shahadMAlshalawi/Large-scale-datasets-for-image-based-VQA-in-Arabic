# Training and Validation Annotations Translation  

```python
import json
with open('original-dataset/mscoco_train2014_annotations.json', 'r') as datafile:
    data = json.load(datafile)
```

```python
import json
with open('original-dataset/mscoco_val2014_annotations.json', 'r') as datafile:
    data = json.load(datafile)
```

```python
data_1 = data['annotations']
```

```python
len(data_1)
```

```python
data_1[0:5]
```

```python
for item in range(0,len(data_1)):
  print(item)
  for an in range(0,len(data_1[item]['answers'])):
    print(an)
    translation_1 = translator.translate(data_1[item]['answers'][an]['raw_answer'], dest='ar')
    translation_2 = translator.translate(data_1[item]['answers'][an]['answer'], dest='ar')
    data_1[item]['answers'][an]['raw_answer_ar'] = translation_1.text
    data_1[item]['answers'][an]['answer_ar'] = translation_2.text
```

```python
import json
with open('annotation_training.json', 'w') as f:
    json.dump(data_1, f, ensure_ascii = False)
```
